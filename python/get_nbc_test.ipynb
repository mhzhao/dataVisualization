```python
import urllib.request

fp = urllib.request.urlopen("https://www.nbcnews.com/")
mybytes = fp.read()

mystr = mybytes.decode("utf8")
fp.close()

print(mystr)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(mystr, 'html.parser')
```

```python
divs = soup.find_all('span')

for d in divs:
    try:
        if len(d['class']) == 1 and d['class'][0] == "headline___38PFH":
            print(d.get_text())
    except:
        pass

# links = soup.find_all('a')

# for a in links:
#     try:
#         if len(a['class']) == 1 and a['class'][0] == "vilynx_disabled":
#             print(a.get_text())
#     except:
#         pass
    
#     title = d.get("data-bbc-title")
#     if title != None:
#         print(title)
```
