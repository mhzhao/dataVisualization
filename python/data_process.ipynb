```python
import pandas as pd
```

```python
code_data = pd.read_csv("../data/GDP/country_code.csv")
gdp_data = pd.read_csv("../data/GDP/API_NY.GDP.MKTP.KD.ZG_DS2_en_csv_v2_713083.csv", skiprows=[0,1,2,3])
```

```python
# gdp_data.fillna(0)
code_data
```

```python
two_list = []
three_list = []
for d in code_data["ISO CODES"].str.split("/"):
    two_list.append(d[0].strip(" "))
    three_list.append(d[1].strip(" "))

three_dict = {}
for index, name in enumerate(code_data["COUNTRY"]):
    three_dict[three_list[index]] = name
    
# three_dict
```

```python
gdp_data["Country/Other"] =  [d in three_dict for d in gdp_data["Country Code"]]
```

```python
country_gdp = gdp_data[gdp_data["Country/Other"]].fillna(0)
```

```python
country_gdp.to_csv("../data/GDP/country_GDP.csv", index=False)
```
